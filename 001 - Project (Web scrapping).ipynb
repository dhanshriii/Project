```python
#import required libraries
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

```python
#We are scrapping the flipkart website so copied the Url of searching
flipkart_url = "https://www.flipkart.com/search?q=" + "Realme"
```

```python
#This will generate the url for that specific product
flipkart_url
```

```python
from urllib.request import urlopen
import ssl

# Ignore SSL certificate verification
urlclient = urlopen(flipkart_url, context=ssl._create_unverified_context())
```

```python
urlclient
```

```python
flipkart_page = urlclient.read()
```

```python
flipkart_page
```

```python
#But this data is not human readable
#lets beutify it
flipkart_html = bs(flipkart_page,'html.parser')
```

```python
bigbox = flipkart_html.findAll("div",{"class" : "_1AtVbE col-12-12"})
```

```python
del bigbox[0:3]
```

```python
product_link = "https://www.flipkart.com" + bigbox[3].div.div.div.a['href']
```

```python
product_req = requests.get(product_link)
```

```python
product_link
```

```python
product_html = bs(product_req.text,'html.parser')
```

```python
comment_box = product_html.find_all("div",{"class" : "_16PBlm"})
```
